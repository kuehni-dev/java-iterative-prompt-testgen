```python
import pandas as pd

import utils
```

```python
PROJECTS = ['Gson', 'JacksonCore', 'Lang']


def read(name: str) -> pd.DataFrame:
    return utils.union_read_csv(*(f'data/{p}_{name}.csv' for p in PROJECTS))


def write_combined(data: pd.DataFrame, name: str, *, index=True) -> None:
    data.to_csv(f'data/out/{name}.csv', index=index)
```

```python
content = read('generation-validity')
sum_row = content.sum().to_frame().T
write_combined(sum_row, 'generation-validity', index=False)
```

```python
content = read('fails-per-iteration-and-feedback-type').groupby(['NUMBER', 'FEEDBACK_TYPE']).sum().unstack().astype('Int64')
content = content[sorted(content.columns, key=lambda c: c[1])]
content.columns = [f'{t}_{m}' for m, t in content.columns]
write_combined(content, 'fails-per-iteration-and-feedback-type')
```
