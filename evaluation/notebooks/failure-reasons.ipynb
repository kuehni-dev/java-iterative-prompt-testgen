```python
import matplotlib.pyplot as plt
import numpy as np

import utils

plt.style.use('default')
```

```python
df = utils.union_read_csv('data/Gson_failure-reasons.csv', 'data/JacksonCore_failure-reasons.csv', 'data/Lang_failure-reasons.csv')
pivot = df.pivot_table(
    index='FEEDBACK_TYPE',
    columns='ERROR',
    values='FREQUENCY',
    aggfunc='sum',
    fill_value=0,
    dropna=False,
).reindex([np.nan, 'COVERAGE', 'MUTATION']).rename(index={
    np.nan: 'Initiale Generierung',
    'COVERAGE': 'Coverage Feedbackschleife',
    'MUTATION': 'Mutation Feedbackschleife',
})
pivot = pivot[pivot.sum(axis=0).sort_values(ascending=False).index]
pivot
```

```python
figsize = utils.A5_LANDSCAPE_INCHES
figsize = (figsize[0], figsize[1] / 3 * 2)

fig, axes = plt.subplots(figsize=figsize, dpi=300)
fig.suptitle('Fehlerhäufigkeiten')

pivot.plot(kind='bar', ax=axes, width=.9)
for cont in axes.containers:
    axes.bar_label(cont, padding=2, fontsize=9)
    for bar in cont:
        x_left = bar.get_x()
        x_right = bar.get_x() + bar.get_width()

        # small vertical extension in data units
        marker_height = axes.get_ylim()[1] * 0.01  # 1% of y-range

        axes.vlines(
            x=[x_left, x_right],
            ymin=0,
            ymax=marker_height,
            linewidth=bar.get_linewidth(),
            color='k',
        )

axes.margins(y=.1)
axes.set_xlabel('')
axes.set_ylabel('Anzahl')
axes.tick_params(axis='x', labelsize=8, rotation=0)
axes.grid(linestyle='--', linewidth=.5, alpha=0.5, axis='y')

# Put legend to the right of the current axis
# axes.legend(title='Fehler', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8)
axes.legend(title='Fehler', loc='upper left', fontsize=8)

plt.show()
fig.savefig('out/failure-reasons.svg', bbox_inches='tight')
```
